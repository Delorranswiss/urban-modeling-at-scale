# 2. Calculate different geometric features from the Geojson files of the buildings, one by one
* **The ideas is simple: Open each building as a polygon (multipolygons) using libraries like Fiona ans Shapely and easily calculate features such as:**
    * Coordinates for the center of the buildings
    * Area of the footprint of each building
    * Number of corners in each building footprint
    * height of the buildings (not available for all)
    * Shape descriptors such as Fourier Descriptor
    * Orientation of each building (along the first and the second axis)
    * Anything else you can imagine to do with building footprints

* **The code below finally will save the building analytics of each country to CSV separate files**

````python
import glob
import os
path = 'path/json_splits/'
all_patches_names = []
counter = 0
for filename in glob.glob(os.path.join(path, '*.geojson')):
    all_patches_names.append(filename)
print ("number of geojson patches for ", region,": ", len(all_patches_names))



from shapely.ops import cascaded_union
import fiona
from shapely.geometry import Polygon
from shapely.geometry import MultiPolygon

all_areas = []
all_no_coords = []
all_xs = [] 
all_ys = []
all_osm_ids = []
all_building_types = []

for patch_no , buildings_path in enumerate(all_patches_names[:]):

    with fiona.open(buildings_path,) as f:
        crs = f.crs
        iterate = 0
        for rec in f:
            if rec['geometry']['type'] == 'MultiPolygon':
                polygons = MultiPolygon([Polygon(rec['geometry']['coordinates'][0][i]) for i in range(len(rec['geometry']['coordinates'][0]))])
                try:
                    areas = cascaded_union(polygons).area
                # areas = [polygons[i].area for i in range(len(polygons))]
                # if len(areas)>1:
                # 	areas =np.asarray(areas)
                # 	areas = np.sort(areas)
                # 	areas = areas[-1]-np.sum(areas[:-1])
                # else:
                # 	areas = areas[0]
                except:
                    print('does not work')
                    areas = 0

            no_coords = [len(polygons[i].exterior.coords)-1 for i in range(len(polygons))]
            no_coords = np.sum(no_coords)
            all_areas.append(areas)
            all_no_coords.append(no_coords)
            all_xs.append(polygons.centroid.xy[0][0])
            all_ys.append(polygons.centroid.xy[1][0])
            # all_osm_ids.append(str(rec['properties']['osm_id']))
            try:
                all_building_types.append(str(rec['properties']['building']))
            except:
                all_building_types.append(str('yes'))


            iterate = iterate +1
    if patch_no%20000==0:
        print ("patch number: ",patch_no,"polygon: ",iterate)
DF =pd.DataFrame()
DF['area'] = all_areas
DF['corners'] = all_no_coords
DF['x_center'] = all_xs
DF['y_center'] = all_ys
# DF['osm_id'] = all_osm_ids
DF['type'] = all_building_types

print ('corner analytics done')
import vaex

buildings_path = "path/buildings/"

csv_path = buildings_path+region+'-latest-building.csv'
DF.to_csv(path_or_buf=csv_path,index=False)
print ('CSV done')
````

## 3.2. Spatial query for each building
* This step (while not very efficient yet) is super simple. We open the building files and we just need the coordinates of the center of each building, previously calculated.
* Then for each POI type (e.g. TRANSPORT_BUSSTOP) or POI category (e.g. TRANSPORT_*) we qualify the buildings.
* Here, we have two options: 
    * 1-To set a radius around each building and count the POIs of the selected category in the given radius
    * 2- To calculate the distance to the closest POI of the selected category. 
- I found the first approach more viable.

# 4. Converting the final CSV files to the maptiles using tippecanoe. This is a great library for huge data sets. 
* https://github.com/mapbox/tippecanoe

### Next step, we will upload this tiles to mapbox for styling our own maps

````python
for key in sel_types:
    import glob
    import os
    
    # Path to qualified buildings for each POI
    path = 'path/OSM_pois_neigh/'
    sel_type = key
    all_csv_names = []
    counter = 1
    for filename in glob.glob(os.path.join(path, '*.csv')):
        #for each POI type and each country there is a unique CSV
        if sel_type in filename:
            all_csv_names.append(filename)

    print (len(all_csv_names))
    import os
    include = sel_type
    output_fn = 'path/tiles/'+sel_type+'.mbtiles'
    layer_n = sel_type
    input_fn =''
    for csv in all_csv_names:
        input_fn = input_fn +' '+ csv
    code  = 'tippecanoe -zg -o {} -l {} --include={} --drop-densest-as-needed --extend-zooms-if-still-dropping {}'.format(output_fn,layer_n,include,input_fn)
    os.system(code)
    print ('done') 
````
